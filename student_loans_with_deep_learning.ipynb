# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

---

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display a sample of y
y[:5]
```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns=["credit_ranking"])

# Review the features DataFrame
X.head()
```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_Train, X_Test, y_Train, y_Test = train_test_split(X, y, random_state=1)
```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the features training dataset
X_Scaler = scaler.fit(X_Train)

# Fit the scaler to the features training dataset
X_Train_Scaled = X_Scaler.transform(X_Train)
X_Test_Scaled = X_Scaler.transform(X_Test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
feature_count = len(X_Train.iloc[0])

# Review the number of features
feature_count
```

```python
# Define the number of hidden nodes for the first hidden layer
firstlayer_nodes = 6

# Define the number of hidden nodes for the second hidden layer
secondlayer_nodes = 3

# Define the number of neurons in the output layer
outputlayer_neurons = 1
```

```python
# Create the Sequential model instance
seq_model = Sequential()

# Add the first hidden layer
seq_model.add(Dense(units=firstlayer_nodes, input_dim=feature_count, activation="relu"))

# Add the second hidden layer
seq_model.add(Dense(units=secondlayer_nodes, activation="relu"))

# Add the output layer to the model specifying the number of output neurons and activation function
seq_model.add(Dense(units=outputlayer_neurons, activation="linear"))
```

```python
# Display the Sequential model summary
seq_model.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
seq_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 50 epochs and the training data
seq_model.fit(X_Train_Scaled, y_Train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = seq_model.evaluate(X_Test_Scaled, y_Test, verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Step 4: Save and export your model to a keras file, and name the file `student_loans.keras`.


```python
# Set the model's file path
file_path = Path("student_loans.keras")

# Export your model to a keras file
seq_model.save(file_path)
```

---
## Predict Loan Repayment Success by Using your Neural Network Model

### Step 1: Reload your saved model.

```python
# Set the model's file path
file_path = Path("student_loans.keras")

# Load the model to a new object
seq_model_loaded = tf.keras.models.load_model(file_path)
```

### Step 2: Make predictions on the testing data and save the predictions to a DataFrame.

```python
# Make predictions with the test data
predictions = seq_model_loaded.predict(X_Test_Scaled, verbose=2)

# Display a sample of the predictions
print(predictions[:5])
```

```python
# Save the predictions to a DataFrame and round the predictions to binary results
predictions_df = pd.DataFrame(predictions)
predictions_df = predictions_df.round(0)
predictions_df
```

### Step 4: Display a classification report with the y test data and predictions

```python
# Print the classification report with the y test data and predictions
print(classification_report(y_Test, predictions_df.values))
```

**1. Describe the data that you would need to collect to build a recommendation system to recommend student loan options for students. Explain why this data would be relevant and appropriate.**
To build a recommendation system for student loans, we need to collect data to capture the financial, educational, and personal circumstances of students, along with their specific loan needs. 
Here is a list of relevant data points that would form the feature set to train the recommendation system:

1. Student Demographic Data - 
 - Age of the student
 - Residency: Domestic or International
 - Location: To determine state specific loans or programs, cost of living.
 - Employment Status: Determines the likelihood of the student’s ability to work part-time to help repay the loan, or the need for income-driven repayment options.
2. Educational Information
 - Program Type: Undergraduate, Graduate, etc.
 - Field of Study
3. Academic Performance
 - GPA or Academic Standing
4. Financial Information
 - Annual Family Income
 - Credit Score, if available
 - Existing Debt or Other Financial Obligations
5. Loan Terms
 - Needed Loan Amount
 - Preferred Repayment Term
 - Interest Rate Type (Fixed vs. Variable)
 - Grace Period for loan repayment
 - Co-signers
6. Scholarships, Grants, and Other Financial Aid
 - Other Financial Aid Sources

Each of the above data points contributes to a holistic understanding of the student’s financial situation, educational goals, and borrowing needs, allowing for personalized recommendations. By combining the data about the student, the system can recommend loan options for students to choose that best fits their needs.

**2. Based on the data you chose to use in this recommendation system, would your model be using collaborative filtering, content-based filtering, or context-based filtering? Justify why the data you selected would be suitable for your choice of filtering method.**
A context-based filtering will be more appropriate to model the student loan recommendation system as it incorporates various contextual data points to provide personalized loan options to meet each student needs.

Context-based filtering leverages the student's financial data to recommend loans that are suitable to the student's situation. It can also recommend loans based on the context of programs or institutions the student may be interested in pursuing and help them narrow down the options. Context-based filtering takes into consideration the demography data of the student and their loan preferences. 

In contrast, collaborative filtering and content-based filtering, will not be as suitable for this application, as the student loan landscape is highly personal and context-dependent. 

By adopting a context-based filtering approach, the student loan recommendation system can provide students with tailored, holistic, and continuously improving suggestions that best meet their individual needs and preferences.

**3. Describe two real-world challenges that you would take into consideration while building a recommendation system for student loans. Explain why these challenges would be of concern for a student loan recommendation system.**
Couple of real-world challenges that we would need to take into account while building a recommendation system are student data privacy and student's financial responsibility. 

Data Privacy: The recommendation system needs student's sensitive financial and personal data to determine their eligibility for various loan options and they must be protected to avoid exposure or misuse. Unintended exposure or mishandling sensitive data may lead to legal consequences, lack of student's trust in the recommendation system and may impact their financial security. The recommendation system should address data privacy challenges by incorporating data encryption, data anonymization and data access controls.

Financial Responsibillity: The recommendation system needs to take into account the student's long term financial well being while recommending loan options. If the system recommends financially burdensome loan option or more debt the student can afford to manage in the long term, it could lead to loan default, and reduce financial mobility after the student graduates. To address the student's long term financial well being, the system should consider risk assessments to evaluate the student's post graduate income based on their field of study and their ability to repay the loan. In addition, the recommendation system should provide clear communication on the implications of loan options so the students can make informed decisions.
